```python
dataset = pd.read_csv('../data/county.csv')
```

```python
dataset.head()
```

```python
dataset.info()
```

```python
dataset.dropna(subset = ['poverty', 'median_hh_income'], inplace=True)
```

```python
dataset.info()
```

```python
X = dataset[['poverty']].values
y = dataset[['median_hh_income']].values
```

```python
# Importing Linear Regression
from sklearn.linear_model import LinearRegression

# Training Model
lm=LinearRegression()
lm.fit(X.reshape(-1,1),y.reshape(-1,1))
```

```python
# importing libraries for polynomial transform
from sklearn.preprocessing import PolynomialFeatures
# for creating pipeline
from sklearn.pipeline import Pipeline
# creating pipeline and fitting it on data
Input=[('polynomial',PolynomialFeatures(degree=2)),('modal',LinearRegression())]
pipe=Pipeline(Input)
pipe.fit(X.reshape(-1,1),y.reshape(-1,1))
```

```python
poly_pred=pipe.predict(X.reshape(-1,1))
#sorting predicted values with respect to predictor
sorted_zip = sorted(zip(X,poly_pred))
x_poly, poly_pred = zip(*sorted_zip)
#plotting predictions
plt.figure(figsize=(12,8))
sns.scatterplot(data=dataset[['poverty', 'median_hh_income']], x="poverty", y="median_hh_income")
plt.plot(x_poly,poly_pred,color='r')
plt.xlabel('poverty')
plt.ylabel('median_hh_income')
```
