```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Random Process
    - Not influenced by other processes or are too complex for us humans to understand
### Outcome
    - Result of a Random process
### Probability
    - Proportion of desired outcomes to all possible outcomes
    - The sum of the probabilities of all outcomes is 1
### Law of large numbers
    - If a random process is run long enough, the probability of an outcome will converge to it's theoretical probability

```python
n, p = 100000, .5  # number of trials, probability of each trial
m = 100000
# result of flipping a coin n times, tested m times.
s = np.random.binomial(n, p, m)
```

```python
df = pd.DataFrame(list(zip(s/n, np.arange(m))), columns = ['P-Hat', 'Trial'])
```

```python
df.head()
```

```python
plt.figure(figsize = (12,8))
sns.scatterplot(data=df, x="Trial", y="P-Hat")
```

### Disjoint outcomes

$P(A_1 or A_2) = P(A_1) + P(A_2)$

## Addition Rule

$P(A_1 \cup A_2) = P(A_1) + P(A_2) - P(A_1 \cap A_2)$

### Probability distibutions



```python
x=np.linspace(2,12,11)
p=np.array([1,2,3,4,5,6,5,4,3,2,1])/36.

plt.figure(figsize = (12,8))

plt.bar(x, p)
plt.xlim(2, 12)
plt.xlabel("Dice Sum")
plt.ylabel("Probability")
```

### Sample space
    - Set of all outcomes of a random process
### Complement
    - Proportion of the undesired outcomes to all possible outcomes
### Independence
    - Two process are independent if the outcome of one provides no useful information about the other

## Multiplication Rule for Indepent Processes

$P(A_1 and A_2) = P(A_1) \times P(A_2)$
